```python
import numpy as np
```

```python
data = np.loadtxt(fname='./inflammation-01.csv',delimiter=",")
```

```python
print(data)
```

```python
import matplotlib.pyplot as plt
```

```python
stdev = np.std(data, axis=0)
stdevplot = plt.plot(stdev)
plt.xlabel('time(days)')
plt.ylabel('standard deviation')
plt.title('standard deviation for each day')
```

```python
ave_inflammation = np.mean(data, axis=0)
ave_plot = plt.plot(ave_inflammation)
plt.xlabel('time(days)')
plt.ylabel('inflammation score')
plt.title ('patient data')
```

```python
man_plot = plt.plot(np.max(data,axis=0))
plt.show()
```

```python
fig = plt.figure(figsize=(10.0,25.0))

sbplt1 = fig.add_subplot(6, 2, 5)
sbplt2 = fig.add_subplot(9,2,5)
sbplt3= fig.add_subplot(15,2,5)

sbplt1.set_xlabel('time(days)')
sbplt1.set_ylabel('average')
sbplt1.plot (np.mean(data,axis=0))

sbplt2.set_ylabel('max')
sbplt2.plot (np.max(data,axis=0))

sbplt3.set_ylabel('min')
sbplt3.plot (np.min(data,axis=0))

fig.tight_layout

plt.show()

```
